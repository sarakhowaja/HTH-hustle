```python
import random

class TicTacToe:

  def __init__(self):
    self.board = ["1","2","3","4","5","6","7","8","9"]
    self.oboard = ["1","2","3","4","5","6","7","8","9"]
    self.current_leader = None
    self.player1 = ""
    self.player2 = ""
    self.position = 0
    self.used_spaces = []
    self.end_game = False
    self.play_again = True
    self.current_winner = None

  def display_board(self):
    print(self.board[0] + "|" + self.board[1] + "|" + self.board[2])
    print("-----")
    print(self.board[3] + "|" + self.board[4] + "|" + self.board[5])
    print("-----")
    print(self.board[6] + "|" + self.board[7] + "|" + self.board[8])

  def player_input(self):
    self.player1 = input("Player 1, choose X or O: ").upper()
    while self.player1 != "X" and self.player1 != "O":
        self.player1 = input("Player 1, choose X or O: ").upper()
    if self.player1 == "X":
      self.player2 = "O"
    else:
      self.player1 == "O"
      self.player2 = "X"
    print(self.player1 + " is player 1")
    print(self.player2 + " is player 2")

  def choose_first(self):
    first = random.randint(1,2)
    if first == 1:
      self.current_leader = self.player1
      print("Player 1 goes first")
    else:
      self.current_leader = self.player2
      print("Player 2 goes first")

  def player_choice(self):
   self.position = int(input("Choose a position (1-9)"))
   while self.position not in range(1,10) or self.position in self.used_spaces:
    self.position = int(input("Choose a position (1-9)"))
   self.used_spaces.append(self.position)

  def place_marker(self):
    self.board[self.position - 1] = self.current_leader
    if self.current_leader == self.player1:
     self.current_leader = self.player2
    else:
     self.current_leader = self.player1

  def board_check(self):
    if len(self.used_spaces) == 9:
      print("It is a Tie.")
    else:
      print("The board is not full.")

  def check_win(self):
    if self.board[0] == self.board[1] and self.board[1] == self.board[2]:
     print(f"Player {self.board[0]} is the winner.")
     self.current_winner = self.board[0]
    elif self.board[3] == self.board[4] and self.board[4] == self.board[5]:
     print(f"Player {self.board[3]} is the winner.")
     self.current_winner = self.board[3]
    elif self.board[6] == self.board[7] and self.board[7] == self.board[8]:
     print(f"Player {self.board[6]} is the winner.")
     self.current_winner = self.board[6]
    elif self.board[0] == self.board[3] and self.board[3] == self.board[6]:
     print(f"Player {self.board[0]} is the winner.")
     self.current_winner = self.board[0]
    elif self.board[1] == self.board[4] and self.board[4] == self.board[7]:
     print(f"Player {self.board[1]} is the winner.")
     self.current_winner = self.board[1]
    elif self.board[2] == self.board[5] and self.board[5] == self.board[8]:
     print(f"Player {self.board[2]} is the winner.")
     self.current_winner = self.board[2]
    elif self.board[0] == self.board[4] and self.board[4] == self.board[8]:
     print(f"Player {self.board[0]} is the winner.")
     self.current_winner = self.board[0]
    elif self.board[2] == self.board[4] and self.board[4] == self.board[6]:
     print(f"Player {self.board[2]} is the winner.")
     self.current_winner = self.board[2]

  def play_game(self):
    while self.end_game == False:
      if len(self.used_spaces) == 0:
        self.player_input()
        self.choose_first()
      self.display_board()
      self.player_choice()
      self.place_marker()
      self.check_win()
      self.board_check()
      if self.current_winner != None:
        choice = input("Do you want to play again? ").upper()
        while choice != "NO" and choice != "YES":
          choice = input("Do you want to play again? ").upper()
        if choice == "NO":
          self.end_game = True
        else:
          self.board = self.oboard
          self.used_spaces.clear()
          self.current_winner = None

game = TicTacToe()
game.play_game()
```
